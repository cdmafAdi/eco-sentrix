```python
input_img = input('Enter name of image file to check if it is water hyacinth')
# input_img = img3
result = CLIENT.infer(input_img, model_id="water-hyacinth-detection-omno0-xqrjy/2")


confidence = result["predictions"][0]["confidence"]
if confidence >= 0.95:
  print('the image detected is of water hyacinth')
  plt.imshow(plt.imread(input_img))
  plt.show()

elif(0,95>confidence>0.90):
  print('the image detected may not be water hyacinth ')
  plt.imshow(plt.imread(input_img))
  plt.show()
else:
  print('the image detected is not water hyacinth')
  plt.imshow(plt.imread(input_img))
  plt.show()
print(confidence)



```
